```python
import random
import matplotlib.pyplot as plt
```

### Optimal Stopping

```python
averages = []
found_bests = []

for stopping_point in range(1, 101):
    final_choices = []
    num_simulations = 100000
    found_best = 0

    for i in range(num_simulations):
        seen_count = 0
        max_seen = 0

        for i, n in enumerate(random.sample(range(1, 101), k=100)):
            seen_count += 1
            if n > max_seen:
                max_seen = n
                if seen_count >= stopping_point:
                    last_seen = max_seen
                    break
            last_seen = n
        if last_seen == 100:
            found_best += 1
        final_choices.append(last_seen)

    averages.append(sum(final_choices)/num_simulations)
    found_bests.append(found_best/num_simulations)
```

```python
max_averages = max(averages)
index_averages = averages.index(max_averages)

max_bests = max(found_bests)
index_bests = found_bests.index(max_bests)
```

```python
fig, ax = plt.subplots(1,2, figsize=(15, 5))
ax[0].plot(averages)
ax[0].set_title("Average outcome")
ax[0].set_xlabel("Optimal stopping value")
ax[0].set_ylabel("Score")
#ax[0].axhline(y=max_averages, xmax=index_averages/100, color='r', linestyle='--')
#ax[0].axvline(x=index_averages, ymax=max_averages/100, color='r', linestyle='--')
ax[0].grid()
ax[1].plot(found_bests)
ax[1].set_title("Probability of finding the best outcome")
ax[1].set_xlabel("Optimal stopping value")
ax[1].set_ylabel("Probability (%)")
ax[1].grid()
plt.show()

print(f"On average, the optimal stopping value is {index_averages}, with an average outcome score of {max_averages}.")
print(f"To maximize the probability of finding the best outcome in the pool, the optimal stopping value is {index_bests}, with a probability of {max_bests}%.")
```

```python
roger_wins = 0
bobby_wins = 0
number_of_games = 10000000
import random
for i in range(0, number_of_games): 
    roger = random.randint(0, 1)
    bobby = random.randint(0, 1)
    if roger > bobby:
        roger_wins += 1
    if roger < bobby:
        bobby_wins += 1
```
