## Signal Processing Study

```python
Fs = 1000 # Sampling frequency                    
T = 1/Fs # Sampling period       
N = 50000 # Number of samples (ms)
t=np.arange(0, N)*T
```

```python
# Form a signal containing a 12 Hz sinusoid of amplitude 1 and a 24 Hz sinusoid of amplitude 0.7
S = 1*np.sin(2*np.pi*12*t) + 0.7*np.sin(2*np.pi*24*t)
```

```python
# Corrupt the signal with zero-mean white noise with a variance of 4
X = S + np.random.normal(loc=0, scale=2, size=N)
```

```python
plt.plot(1000*t[0:50], X[0:50])
plt.title("Time Domain  - Signal Corrupted with Zero-Mean Random Noise")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude")
plt.show()
```

scipy.fftpack.fft is an implementation of the Fast Fourier Transform (FFT), which is an efficient algorithm for calculating the Discrete Fourier Transform (DFT).

If you pass in a time-domain signal of length  N  scipy.fftpack.fft, you will get back a frequency-domain signal of length  N .

The first  N2  samples in the frequency domain represent the DFT values at equally spaced positive frequencies in  [0,fs/2) .
The last  N2  samples represent the DFT values at equally spaced negative frequencies in  [fs/2,0) .
If the time-domain signal is real-valued, then the DFT values at negative frequencies will be the complex conjugates of the DFT values at positive frequencies. So all of the information is contained in the first half of the DFT for a real-valued signal.

```python
# Compute the Fourier transform of the signal
Y = fft(X)
# Compute the two-sided spectrum P2. Then compute the single-sided spectrum P1 based on P2 and the even-valued signal length
P2 = np.abs(Y/N)
P1 = 2*P2[0:int(N/2)]
#Define the frequency domain f
f = np.arange(0, Fs/2, step=Fs/N)
```

```python
# Plot the single-sided amplitude spectrum P1. 
# The amplitudes are not exactly at 1 and 0.7, as expected, because of the added noise. 
# On average, longer signals produce better frequency approximations.
plt.plot(f, P1)
plt.title("Frequency Domain  - Signal Corrupted with Zero-Mean Random Noise")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Find peaks inside the signal based on height condition
peaks = scipy.signal.find_peaks(P1, height=0.2)
peaks
```

```python
# Find peak frequency
peaks[0]*Fs/N
```

```python
# Define a function to return the best model score and parameters
def best_model(model):
    print(model.best_params_)
    print(model.best_estimator_)
    print("Train Score: ", model.best_score_)
    
def get_auc_scores(y_actual, y_pred, y_pred_proba):
    auc_score = roc_auc_score(y_actual, y_pred); 
    fpr, tpr, _ = roc_curve(y_actual, y_pred_proba); 
    return (auc_score, fpr, tpr)
```
