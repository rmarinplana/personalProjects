```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seattle_weather = pd.read_csv("../datasets/visualizations/seattle_weather.csv")
austin_weather = pd.read_csv("../datasets/visualizations/austin_weather.csv")
```

```python
# Create a figure and an array of axes: 2 rows, 1 column with shared y axis
fig, ax = plt.subplots(2, 1, sharey=True)

# Plot Seattle temperature data in the top axes
ax[0].plot(seattle_weather["DATE"], seattle_weather["MLY-TAVG-NORMAL"], color = "b")

# Plot Austin temperature data in the bottom axes
ax[1].plot(austin_weather["DATE"], austin_weather["MLY-TAVG-NORMAL"], color = "r")

plt.show()
```
